```python
fd = open('input.txt').readlines()
```

```python
scores = {
    'A': 1,
    'B': 2,
    'C': 3,
    'X': 1,
    'Y': 2,
    'Z': 3,
}
```

```python
from tqdm import tqdm

total_score = 0

for i in tqdm(range(len(fd))):
    el1 = fd[i].split(' ')[0]
    el2 = fd[i].split(' ')[1].split('\n')[0]

    if el1 == 'A' and el2 == 'X':
        match_score = 3
    elif el1 == 'A' and el2 == 'Y':
        match_score = 6
    elif el1 == 'A' and el2 == 'Z':
        match_score = 0
    elif el1 == 'B' and el2 == 'X':
        match_score = 0
    elif el1 == 'B' and el2 == 'Y':
        match_score = 3
    elif el1 == 'B' and el2 == 'Z':
        match_score = 6
    elif el1 == 'C' and el2 == 'X':
        match_score = 6
    elif el1 == 'C' and el2 == 'Y':
        match_score = 0
    elif el1 == 'C' and el2 == 'Z':
        match_score = 3


    total_score += match_score + scores[el2]

print(total_score)
```
